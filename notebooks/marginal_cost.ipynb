# Visualize Marginal Cost


```python
from cdxbasics.kwargs import dctkwargs
from cdxbasics.logger import Logger
_log = Logger("Log")
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
from scipy.stats import norm

print("/Done")
```

```python
x = np.linspace(-2.,2.,51)
y = np.abs(x)
y = y+0.5*y**2
fig = plt.figure(figsize=(8,6))

plt.title("Cost and marginal cost")
plt.plot(x,y,label="c(a):=$|a|+0.5 a^2$")
plt.plot(x,np.abs(x),label="m(a):=$|a|$")
plt.legend()
plt.show()
```
